```python
import pandas as pd
data=pd.read_csv('train_ctrUa4K.csv')
```

```python
data1=data.drop('Loan_ID',axis=1)
```

```python
data1['Property_Area']=data1['Property_Area'].replace({'Rural':1,'Urban':2,'Semiurban':3})
data1['Self_Employed']=data1['Self_Employed'].replace({'Yes':0,"No":1})
data1['Married']=data1['Married'].replace({'Yes':0,'No':1})
data1['Dependents']=data1['Dependents'].replace({'3+':3})
data1['Education']=data1['Education'].replace({'Graduate':8,'Not Graduate':9})
data1['Gender']=data1['Gender'].replace({'Male':4,'Female':5})
```

```python
data1=data1.fillna(0)
data1.head(10)
```

```python
#Target Imbalance 
#UPSampling LOAN_STATUS VALS i.e. (No) to 400 vals
from sklearn.utils import resample
```

```python
df_majority=data1[data1['Loan_Status']=='Y']
df_minority=data1[data1['Loan_Status']=='N']

df_minority_upsampled=resample(df_minority,
                               replace=True,
                               n_samples=400,
                               random_state=22)
df_sampled=pd.concat([df_majority,df_minority_upsampled])
df_sampled.Loan_Status.value_counts()
```

```python
df_sampled.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

X1=df_sampled.drop('Loan_Status',axis=1)
y1=df_sampled['Loan_Status']

clf_lr=LogisticRegression().fit(X1,y1)
pred_y_1=clf_lr.predict(X1)
print(accuracy_score(y1,pred_y_1))
```
